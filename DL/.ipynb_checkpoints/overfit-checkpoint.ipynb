```python
import torch
import os

device = torch.device("cuda") if torch.cuda.is_available() else torch.device("cpu")
```

```python
import torchvision
from torchvision import transforms

train_dataset = torchvision.datasets.FashionMNIST(
    root = './data/FashionMNIST',
    train = True,
    download = True,
    transform = transforms.Compose([
        transforms.ToTensor(),
        transforms.Normalize([0.], [0.5])                               
    ])
)

test_dataset = torchvision.datasets.FashionMNIST(
    root = './data/FashionMNIST',
    train = False,
    download = True,
    transform = transforms.Compose([
        transforms.ToTensor(),
        transforms.Normalize([0.], [0.5])                             
    ])
)
```

```python
n_train, n_test = len(train_dataset), len(test_dataset)
```

```python
BATCH_SIZE = 128
NUM_WORKERS = 4
```

```python
train_loader = torch.utils.data.DataLoader(train_dataset, 
                                           batch_size=BATCH_SIZE,
                                           shuffle=True,
                                           num_workers=NUM_WORKERS)
```

```python
test_loader = torch.utils.data.DataLoader(test_dataset,
                                          batch_size=BATCH_SIZE,
                                          num_workers=NUM_WORKERS)
```

# First part

```python
import torch.nn as nn
from models import model_cnn # simple model with Conv, Linear, Relu modules

model_cnn = model_cnn.to(device)
optimizer = torch.optim.Adam(model_cnn.parameters(), lr=3e-4)
criterion = nn.CrossEntropyLoss()
criterion = criterion.to(device)
```

## Структура простой модели без BN, Dropout и т.д.: 

```python
model_cnn
```

```python
import time
from tqdm import tqdm_notebook as tqdm
import numpy as np

def train_model(model, train_loader, test_loader, optimizer, criterion, n_epochs, top_test_accuracy = 92):

    train_loss_history = []
    test_loss_history = []
    train_accuracy_history = []
    test_accuracy_history = []
    

    for epoch in range(n_epochs):
        start_time = time.time()

        train_loss = []
        test_loss = []
        test_accuracy = []
        train_accuracy = []

        model.train(True) 
        for (X_batch, y_batch) in tqdm(train_loader, desc='Training'):
            X_batch, y_batch = X_batch.to(device), y_batch.to(device)

            optimizer.zero_grad()

            y_batch_pred = model(X_batch)
            loss = criterion(y_batch_pred, y_batch)
            loss.backward()
            optimizer.step()

            train_loss.append(loss.cpu().data.numpy())

            _, y_batch_pred = torch.max(y_batch_pred, 1)
            train_accuracy.append(np.mean( (y_batch.cpu() == y_batch_pred.cpu()).numpy() ))
       
        model.train(False)
        for X_batch, y_batch in test_loader:
            X_batch, y_batch = X_batch.to(device), y_batch.to(device)

            y_batch_pred = model(X_batch)
            loss = criterion(y_batch_pred, y_batch)
            test_loss.append(loss.cpu().data.numpy())

            _, y_batch_pred = torch.max(y_batch_pred, 1)

            test_accuracy.append(np.mean( (y_batch.cpu() == y_batch_pred.cpu()).numpy() ))

            
        # print the results for this epoch:
        print("Epoch {} of {} took {:.3f}s".format(
              epoch + 1, n_epochs, time.time() - start_time))
        
        train_loss_value = np.mean(train_loss)
        test_loss_value = np.mean(test_loss)
        test_accuracy_value = np.mean(test_accuracy) * 100
        train_accuracy_value = np.mean(train_accuracy) * 100

        train_loss_history.append(train_loss_value)
        test_loss_history.append(test_loss_value)
        train_accuracy_history.append(train_accuracy_value)
        test_accuracy_history.append(test_accuracy_value)


        print("  training loss (in-iteration): \t{:.6f}".format(train_loss_value))
        print("  test accuracy: \t\t\t{:.2f} %".format(test_accuracy_value))

        if test_accuracy_value >= top_test_accuracy:
            top_test_accuracy = test_accuracy_value
            best_model = model

    history = (train_loss_history, test_loss_history, train_accuracy_history, test_accuracy_history)
    return best_model, history 
```

```python
NUM_EPOCHS = 15
best_model, history = train_model(model_cnn, train_loader, test_loader, optimizer, criterion, n_epochs = NUM_EPOCHS)
```

# Вывод:

### Как мы видим после 15 эпох мы получили test accuracy > 92%, что и требовалось.

# Second part

```python
import matplotlib.pyplot as plt

def show_history(history):
    epochs = np.arange(1, NUM_EPOCHS + 1)
    train_loss, test_loss, train_acc, test_acc = history
    fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12,5))

    for i, (train, test, ylabel) in enumerate(zip((train_loss, train_acc), (test_loss, test_acc), ('Loss', 'Accuracy'))):
        axes[i].plot(epochs, train, label = 'train')
        axes[i].plot(epochs, test, label = 'test')
        axes[i].set_xlabel('Number of epoch')
        axes[i].set_ylabel(ylabel)
        axes[i].grid(True)
        axes[i].legend(loc="lower left")
        
    plt.show();
```

```python
from models import model_cnn_overfit

model_cnn_overfit = model_cnn_overfit.to(device)

optimizer = torch.optim.Adam(model_cnn_overfit.parameters(), lr=3e-4)
criterion = nn.CrossEntropyLoss()
criterion = criterion.to(device)
```

## Структура сложной модели(по сравнению с простой добавлено несколько слоев) без BN, Dropout и т.д.: 

```python
model_cnn_overfit
```

```python
NUM_EPOCHS = 20
_, history_overfit = train_model(model_cnn_overfit, train_loader, test_loader, optimizer, criterion, n_epochs = NUM_EPOCHS)
```

```python
show_history(history_overfit)
```

# Вывод:

### По графикам видим, как модель переобучилась( так как loss на test быстро растет). Что и требовалось получить.
